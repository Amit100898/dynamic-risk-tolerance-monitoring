```python
df = pd.read_csv(
    "../data/lending_club.csv",
    usecols=use_cols,
    nrows=200000,
    low_memory=False
)
```

```python
df.info()
```

```python
df.describe()
```

```python
# Fill numeric missing values with median
num_cols = df.select_dtypes(include=["float64"]).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python

# Fill categorical missing values with mode
cat_cols = df.select_dtypes(include=["object", "string"]).columns

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.isnull().sum()
```

```python
df["term"] = df["term"].str.replace(" months", "").astype(int)
```

```python
df["emp_length"] = (
    df["emp_length"]
    .str.replace("+", "", regex=False)
    .str.replace(" years", "", regex=False)
    .str.replace(" year", "", regex=False)
    .str.replace("< 1", "0")
    .astype(float)
)
```

```python
df["issue_d"] = pd.to_datetime(df["issue_d"], format="%b-%Y")

```

```python
df.info()
```

```python
df["month"] = df["issue_d"].dt.to_period("M").astype(str)
```

```python
df["fico_avg"] = (df["fico_range_low"] + df["fico_range_high"]) / 2

df["credit_utilization"] = df["revol_util"] / 100

df["loan_income_ratio"] = df["loan_amnt"] / df["annual_inc"]

df["delinq_flag"] = (df["delinq_2yrs"] > 0).astype(int)
```

```python
from sklearn.preprocessing import MinMaxScaler

risk_features = [
    "fico_avg",
    "dti",
    "int_rate",
    "credit_utilization",
    "loan_income_ratio",
    "delinq_flag"
]

scaler = MinMaxScaler()

df_scaled = pd.DataFrame(
    scaler.fit_transform(df[risk_features]),
    columns=risk_features
)
```

```python
(df["annual_inc"] == 0).sum()
```

```python
df["annual_inc"] = df["annual_inc"].replace(0, np.nan)
```

```python
df["loan_income_ratio"] = df["loan_amnt"] / df["annual_inc"]
```

```python
df["loan_income_ratio"].replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
df["loan_income_ratio"].fillna(df["loan_income_ratio"].median(), inplace=True)
```

```python
df_scaled = pd.DataFrame(
    scaler.fit_transform(df[risk_features]),
    columns=risk_features
)
```

```python
df = df.replace([np.inf, -np.inf], np.nan)
df = df.fillna(df.median(numeric_only=True))
```

```python
df["risk_score"] = (
    0.30 * (1 - df_scaled["fico_avg"]) +
    0.25 * df_scaled["dti"] +
    0.20 * df_scaled["int_rate"] +
    0.10 * df_scaled["credit_utilization"] +
    0.10 * df_scaled["delinq_flag"] +
    0.05 * df_scaled["loan_income_ratio"]
) * 100
```

```python
df["risk_score"].describe()
```

```python
def risk_tolerance(score):
    if score < 30:
        return "High Risk Tolerance"
    elif score < 60:
        return "Moderate Risk Tolerance"
    else:
        return "Low Risk Tolerance"

df["risk_tolerance"] = df["risk_score"].apply(risk_tolerance)
```

```python
df[["risk_score", "risk_tolerance"]].head()
```

```python
df["risk_tolerance"].value_counts(normalize=True) * 100
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(7,5))
sns.countplot(x="risk_tolerance", data=df)
plt.title("Risk Tolerance Distribution")
plt.xlabel("Risk Category")
plt.ylabel("Number of Borrowers")

plt.savefig("../outputs/charts/risk_distribution.png")
plt.show()
```

```python
df["month"] = df["issue_d"].dt.to_period("M").astype(str)
```

```python
monthly_risk = (
    df.groupby("month")["risk_score"]
      .mean()
      .reset_index()
)

monthly_risk["month"] = pd.to_datetime(monthly_risk["month"])
monthly_risk = monthly_risk.sort_values("month")
```

```python
plt.figure(figsize=(12,6))
plt.plot(monthly_risk["month"], monthly_risk["risk_score"])
plt.title("Dynamic Risk Tolerance Monitoring")
plt.xlabel("Time")
plt.ylabel("Average Risk Score")
plt.grid(True)

plt.savefig("../outputs/charts/dynamic_risk_trend.png")
plt.show()
```
